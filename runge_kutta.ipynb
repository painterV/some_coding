```python
import math

n = 100
nt = 10
m = 5;
def main():
    y1 = [0.0] * (n+1)
    y2 = [0.0] * (n+1)
    y = [0.0] * 2
    # Set up time step and initial values
    dt = 3*math.pi/nt
    y1[0] = 0
    y[0] = 0
    y2[0] = 2
    y[1] = 2
    # Perform the 4th-order Runge-Kutta integration
    for i in range(n):
      t = dt*i
      y = rungeKutta(y, t, dt)
      y1[i+1] = y[0]
      y2[i+1] = y[1]
      #Bring theta back to the region [-pi, pi]
      np =  int(y1[i+1]/(2*math.pi)+0.5)
      y1[i+1] -= 2*math.pi*np
    # Output the result in every m time steps
    for i in range(0, n+1, m):
      print("Angle: ", y1[i])
      print("Angular velocity: ", y2[i], '\n')
# Method to complete one Runge-Kutta step.
def rungeKutta(y, t, dt):
    l = len(y)
    c1 = [0.0] * l
    c2 = [0.0] * l
    c3 = [0.0] * l
    c4 = [0.0] * l
    c1 = g(y, t);
    for i in range(l):
        c2[i] = y[i] + dt*c1[i]/2
    c2 = g(c2, t+dt/2)
    for i in range(l):
        c3[i] = y[i] + dt*c2[i]/2
    c3 = g(c3, t+dt/2)
    for i in range(l):
        c4[i] = y[i] + dt*c3[i]
    c4 = g(c4, t+dt)
    for i in range(l):
      c1[i] = y[i] + dt*(c1[i]+2*(c2[i]+c3[i])+c4[i])/6;
    return c1
#Method to provide the generalized velocity vector.
def g(y, t):
    l = len(y)
    q = 0.5
    b = 0.9
    omega0 = 2.0/3
    v = [0.0] * l
    v[0] = y[1]
    v[1] = -math.sin(y[0])+b*math.cos(omega0*t)
    v[1] -= q*y[1]
    return v

if __name__ == '__main__':
    main()
```
